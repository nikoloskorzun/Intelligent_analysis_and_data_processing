## 1. модель 1 полносвязная сеть
со слоями ... 
- Input(shape=(input_shape,)
- Dense(du, activation='relu') * НЕКОТОРОЕ КОЛИЧЕЕСТВО
- Dense(num_classes, activation='softmax')


## 2. одномерная сверточная сеть(1_d_cnn)
Со слоями ...
- Input(shape=(input_shape, 1))
- Conv1D(filters=f, kernel_size=kernel_size, activation='relu', padding='same') * НЕКОТОРОЕ КОЛИЧЕЕСТВО
- ВОЗМОЖНО BatchNormalization()
- Flatten()     
- Dense(du, activation='relu') * НЕКОТОРОЕ КОЛИЧЕЕСТВО
- ВОЗМОЖНО BatchNormalization()
- Dense(num_classes, activation='softmax')        
## 3. Классификатор в основе которого лежит градиентый бустинг.
